### https://adventofcode.com/2024/day/9

```python
# Reading the data input file
input_file = open("challenge_9_input.txt", "r")
disk_map = []
id_counter = 0
checksum = 0

# The input line is read in as a list of ints
for line in input_file:
    input_data = line.strip()
    input_data =  list(map(int, input_data))

# Out of the input, we create a disk map according to the specifications of the challenge
for index,value in enumerate(input_data):
    # If the index is even, also including zero, the id number is added to the disk space in the same amount as 
    # the value at this index of the input list.
    if index % 2 == 0:
        for i in range(value):
            disk_map.append(str(id_counter))
        id_counter += 1
    # If it is odd, dots are inserted instead in the same amount as the value at the index of the input data
    # list
    else:
        for i in range(value):
            disk_map.append('.')
            
# We start with the starting index of the list and the last
left_to_right_index = 0
right_to_left_index = len(disk_map) - 1

# We find the first value from the left that is a dot.
while disk_map[left_to_right_index] != '.':
    left_to_right_index += 1

# we find the first value from the right that is not a dot
while disk_map[right_to_left_index] == '.':
    right_to_left_index -= 1

# We iterate through the disk map with while loops because modifications are made to the disk
# map underway.
while left_to_right_index < right_to_left_index:
    
    # The value at the right most of the list is inserted at the left most dot
    disk_map[left_to_right_index] = disk_map[right_to_left_index]
    # the value tha was at the right most of the list is replaced with the dot tha was a the left most
    # number of the list
    disk_map[right_to_left_index] = '.'
        
    while disk_map[left_to_right_index] != '.':
        left_to_right_index += 1
    
    while disk_map[right_to_left_index] == '.':
        right_to_left_index -= 1

# We calculate the checksum
for index,value in enumerate(disk_map):
    # as soon as we find a dot, we stop because there are no more number in the disk map
    if(value == '.'):
        break
    # Otherwise, we multiply the value of the disk map with the index at which the value is and add that number
    # to the checksum
    checksum += (int(value) * index)
    
print(checksum)
```

## Part 2

```python
# This function takes a list and finds the first segment in the list that has '.' as a value
def find_first_free_disk_space_from_left(disk_map = [], indexes = []):
    # As long as a dot is not found, we keep moving the index that will be the start of the segment
    while disk_map[indexes[0]] != '.':
        indexes[0] += 1

    # We start the index that will be the end of the segment at the start of the segment
    indexes[1] = indexes[0]

    # We take a step in the list as long as the next value is a dot
    while disk_map[indexes[1] + 1] == '.':
        indexes[1] += 1
    
    # We return the two indexes, which are the segment
    return indexes
```

```python
# This function finds a segment of equal numbers starting at a provided index, and iterates down a list
def find_next_right_most_file_block(disk_map = [], indexes = []):
    # The starting index is the ending index of the previous, one step down
    # The assumption is the the previous ending index ended a segment
    indexes[0] = indexes[1] - 1
    
    # As long as the value at the index is a dot, we go down the list to find
    # the first number
    while disk_map[indexes[0]] == '.':
        indexes[0] -= 1
    
    # when that number is found, what will be the ending index is placed at the starting index
    indexes[1] = indexes[0]
    
    # We go down the list for as long as the preceding value is the same as the value of the starting index
    while disk_map[indexes[1] - 1] == disk_map[indexes[0]]:
        indexes[1] -= 1
     
    # We return the indexes which represent the segment
    return indexes
```

```python
# This function works in the same way as the previous function, but from the left
# the only difference is that is as a guard to prevent the while loop from going out
# of bound
def find_next_free_disk_space_from_left(disk_map = [], indexes = []):
    indexes[0] = indexes[1] + 1
    
    while disk_map[indexes[0]] != '.':
        indexes[0] += 1
    
    indexes[1] = indexes[0]
    
    # the ending index will never be bigger then the last possible index in the list
    while indexes[1] < len(disk_map) - 1 and disk_map[indexes[1] + 1] == '.':
        indexes[1] += 1
        
    return indexes
```

```python
# Reading the data input file
input_file = open("challenge_9_input.txt", "r")
disk_map = []
id_counter = 0
checksum = 0

# We transform the string into a list of ints
for line in input_file:
    input_data = line.strip()
    input_data =  list(map(int, input_data))

# We create the disk_map according to the specifications
for index,value in enumerate(input_data):
    if index % 2 == 0:
        for i in range(value):
            disk_map.append(str(id_counter))
        id_counter += 1
    else:
        for i in range(value):
            disk_map.append('.')
            
left_to_right_indexes = [0,0]
right_to_left_indexes = [len(disk_map) - 1,len(disk_map) - 1]

# We find the first free disk space on the left
left_to_right_indexes = find_first_free_disk_space_from_left(disk_map, left_to_right_indexes)

# We find the first file starting from the right
while disk_map[right_to_left_indexes[1] - 1] == disk_map[right_to_left_indexes[0]]:
    right_to_left_indexes[1] -= 1

# This double while loop is there so that, for each file found from the right, we look through the whole list starting from the 
# left to see if there is an empty space big enough to contain the file
# The outer most while loop iterates through the files from the right of the list
while left_to_right_indexes[1] < right_to_left_indexes[1]:
    
    # This while loop iterates through all the empty disk spaces of the list starting from the left, for each file found
    # This loop will never look for an empty disk space passed the current file
    while left_to_right_indexes[1] < right_to_left_indexes[1]:
        
        # If the current empty space big enough is found, we move the file into the empty space. 
        # If the file does not take the whole space the remaining empty space will be open to 
        # receive a smaller file in further iterations
        if (left_to_right_indexes[1] - left_to_right_indexes[0] + 1) >= (right_to_left_indexes[0] - right_to_left_indexes[1] + 1):
            for index in range(left_to_right_indexes[0], left_to_right_indexes[0] + (right_to_left_indexes[0] - right_to_left_indexes[1] + 1)):
                disk_map[index] = disk_map[right_to_left_indexes[0]]
            for index in range(right_to_left_indexes[1], right_to_left_indexes[0] + 1):
                disk_map[index] = '.'
            # We break out of the inner loop, to stop looking for more empty file spaces
            break
        
        # We iterate to the next empty disk space if the current empty space is not big enough
        left_to_right_indexes = find_next_free_disk_space_from_left(disk_map, left_to_right_indexes)

    # We move to the next file, either if a big enough empty space found or we did not find any empty space
    right_to_left_indexes = find_next_right_most_file_block(disk_map, right_to_left_indexes)

    # We reset the indexes for the empty spaces
    left_to_right_indexes = [0,0]
    # And find the first one again
    left_to_right_indexes = find_first_free_disk_space_from_left(disk_map, left_to_right_indexes)

# We calculate the checksum according to the specification
# Because the dots can now be scattered through the list, we go through the whole list no matter what
# and add to the checksum when it is not a dot
for index,value in enumerate(disk_map):
    if(value != '.'):
        checksum += (int(value) * index)
    
print(checksum)
```
